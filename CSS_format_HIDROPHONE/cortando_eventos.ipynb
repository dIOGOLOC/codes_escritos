## TRIM DATA EVENT

```python
CUT_BEFORE_P = 30
CUT_AFTER_P = 1800
```

```python
kstnm = ['H2','H4','H5']
stla = [1.3297,0.4123,0.1552] 
stlo = [-31.3445,-24.6437,-27.7875]
event_data_folder = ['/home/diogo/dados_doutorado/HIDROPHONE/H2/','/home/diogo/dados_doutorado/HIDROPHONE/H4/','/home/diogo/dados_doutorado/HIDROPHONE/H5/']
```

```python
for i,j in enumerate(event_data_folder):
    lst_file = glob.glob(j+'*.sac')
    for x,c in enumerate(lst_file):
        #check DATETIME
        st = op.read(c,format='SAC')
        for k,l in enumerate(dic_event['ev_julday']):
            t = op.UTCDateTime(dic_event['ev_timeUTC'][k])

            if st[0].stats.starttime.datetime < t.datetime < st[0].stats.endtime.datetime:

                #st.decimate(factor=10, strict_length=False)
                print(st)
                
                dist,az,baz = op.geodetics.gps2dist_azimuth(float(dic_event['evla'][k]),float(dic_event['evlo'][k]),stla[i],stlo[i])
                gcarc = op.geodetics.kilometer2degrees(dist/1000)
                
                if dist/1000 < 500:
                    try:
                        #Calculating time P
                        model = TauPyModel(model="ak135")
                        arrivals = model.get_travel_times(source_depth_in_km=float(dic_event['evdp'][k]), distance_in_degree=gcarc, phase_list=["P",'Pg'])
                        arr = arrivals[0]

                        #Reference time
                        starttime = (t+arr.time)-CUT_BEFORE_P
                        endtime = (t+arr.time)+CUT_AFTER_P

                        print('P = '+str(arr.time-CUT_BEFORE_P))

                        #Calculating time Pn
                        arrivals_PN = model.get_travel_times(source_depth_in_km=float(dic_event['evdp'][k]), distance_in_degree=gcarc, phase_list=["Pn"])
                        arr_PN = arrivals_PN[0]

                        time_PN = arr_PN.time - (arr.time - CUT_BEFORE_P) 
                        print('Pn = '+str(time_PN))

                        #Calculating time Pg
                        arrivals_PG = model.get_travel_times(source_depth_in_km=float(dic_event['evdp'][k]), distance_in_degree=gcarc, phase_list=["Pn"])
                        arr_PG = arrivals_PG[0]

                        time_PG = arr_PG.time - (arr.time - CUT_BEFORE_P) 
                        print('Pg = '+str(time_PG))

                        if st[0].stats.starttime.datetime < (t+arr.time)-CUT_BEFORE_P < (t+arr.time)+CUT_AFTER_P < st[0].stats.endtime.datetime:

                            #Creating Event Directory 
                            event_directory = DIR_EVENT+kstnm[i]+'/'+'{:04}'.format(starttime.year)+'/'+'{:03}'.format(starttime.julday)+'/'+'{:04}'.format(starttime.year)+'.'+'{:03}'.format(starttime.julday)+'.'+'{:02}'.format(starttime.hour)+'.'+'{:02}'.format(starttime.minute)+'.'+'{:02}'.format(starttime.second)+'.'+'{:03}'.format(starttime.microsecond)[:3]
                            os.makedirs(event_directory, exist_ok=True)

                            st[0].trim(starttime,endtime)
                            print(st[0])


                            headerHHZ = {
                                        'kstnm': kstnm[i], 'kcmpnm': 'HHZ', 'stla': float(stla[i]), 'stlo': float(stlo[i]), 'evdp': float(dic_event['evdp'][k]),  
                                        'evla': float(dic_event['evla'][k]), 'evlo': float(dic_event['evlo'][k]), 'mag': float(dic_event['mag'][k]),'nzjday': int(starttime.julday), 
                                        'nzhour': int(starttime.hour), 'nzmin': int(starttime.minute), 'nzmsec': int('{:03}'.format(starttime.microsecond)[:3]),'nzsec': int(starttime.second),
                                        'nzyear': int(starttime.year),	'cmpaz': 0.0, 'cmpinc': 0.0, 'dist': float(dist/1000), 'gcarc': float(gcarc), 
                                        'az': float(az), 'baz': float(baz), 'user8': float(arr.ray_param/6371),'o':float(CUT_BEFORE_P),'delta':st[0].stats.delta,'t1':float(time_PN),'t2':float(time_PG)
                                            }
                            sacHHZ = op.io.sac.sactrace.SACTrace(data=st[0].data, **headerHHZ)
                            sacHHZ.write(event_directory+'/'+kstnm[i]+'.'+'{:04}'.format(starttime.year)+'.'+'{:03}'.format(starttime.julday)+'.'+'{:02}'.format(starttime.hour)+'.'+'{:02}'.format(starttime.minute)+'.'+'{:02}'.format(starttime.second)+'.'+'{:03}'.format(starttime.microsecond)[:3]+'.sac')
                    except:
                        print("Problema")
```
