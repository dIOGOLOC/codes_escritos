```python
trace = st_hhz[0]

tdata = trace.data
ttime = np.linspace(start=st_hhz[0].times('matplotlib')[0],stop=st_hhz[0].times('matplotlib')[-1],num=len(tdata))

dt = trace.stats.delta
npts = trace.stats.npts
samp_rate = trace.stats.sampling_rate
time_slice = int(trace.stats.npts/samp_rate)

wave_to_plot = np.array_split(tdata,time_slice)
time_to_plot = np.array_split(ttime,time_slice)


# set up plotting
fig = plt.figure()
ax = plt.axes()

img = plt.imread("/home/diogoloc/dados_posdoc/ON02_analysis/Figuras/Localizacao_navio_ponte_editado.jpg")
ax.imshow(img,aspect='auto', extent=[ttime[0],ttime[-1], tdata.min(),tdata.max()])

ax.plot(ttime,tdata,c='k',lw=0.1,alpha=0.3)
line, = ax.plot([],[],c='r', lw=0.5,alpha=0.7)
ax.set_xlim(ttime[0],ttime[-1])
ax.set_ylim(tdata.min(), tdata.max())


ax.set_title('Data: '+st_hhz[0].stats.starttime.strftime('%d/%m/%y  s %H:%M:%S'), loc='left', y=0.85, x=0.02, fontsize='medium')

ax.set_yticks([])
ax.set_xlabel('Tempo (s), 20x mais rpido que o normal')

min_loc = mdates.SecondLocator(interval=20)
min_loc_fmt = mdates.DateFormatter('%M:%S')
ax.xaxis.set_major_locator(min_loc)
ax.xaxis.set_major_formatter(min_loc_fmt)

sec_loc = mdates.SecondLocator(interval=1)
ax.xaxis.set_minor_locator(sec_loc)

# Rotates and right-aligns the x labels so they don't crowd each other.
for label in ax.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')

# methods for animation
def init():
    line.set_data([],[])
    return line,

def animate(i):
    y = wave_to_plot[i]
    x = time_to_plot[i]
    line.set_data(x,y)

    return line,

# Save the animation
anim = animation.FuncAnimation(fig, animate, init_func=init,frames=len(wave_to_plot))
HTML(anim.to_jshtml())

# Save the animation as an 
f = r"/home/diogoloc/dados_posdoc/ON02_analysis/Figuras/Localizacao_navio_ponte_ON_ON02.mp4" 
writervideo = animation.FFMpegWriter(fps=10) 
anim.save(f, writer=writervideo)
```
