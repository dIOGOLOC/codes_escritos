## Removing the instrumental response


```python
for i,j in enumerate(sta_list):
    print('Evento - '+j.split('/')[8])
    st = op.read(j)

    #selecting traces whose channel is *channel*
    channel = j.split('.')[3]
    print("channel para a estação: "+channel)

    traces = [t for t in st if t.stats['channel'] == channel]

    # getting unique station names
    stations = set(t.stats['station'] for t in traces)

    # getting network, first day and last day of each station
    stationstats = {}
    for stationname in stations:
        # traces of station
        stationtraces = [t for t in traces if t.stats['station'] == stationname]

        # network of station
        networks = set(t.stats['network'] for t in stationtraces)
        network = list(networks)[0]

        # first and last day of data
        firstday = min(t.stats['starttime'].date for t in stationtraces)
        lastday = max(t.stats['endtime'].date for t in stationtraces)

        # appending stats
        stationstats[stationname] = {
                'network': network,
                'firstday': firstday,
                'lastday': lastday
        }

    inventories = []

    # list of *.dataless files
    flist = glob.glob(pathname=os.path.join(dataless_dir, "*.dataless"))

    print("Dataless para a estação: ")
    print([stationstats[stationname]['network'], stationname])

    for f in flist:
            inv = Parser(f)
            inventories.append(inv)

    for inv in inventories:
            if stationname in inv.get_inventory()['stations'][0]['station_id']:
                    paz = inv.get_paz(channel)
                    stla = inv.get_coordinates(channel)['latitude']
                    stlo = inv.get_coordinates(channel)['longitude']
    print('Coordenadas:',stla,stlo)
    st_orig = st.copy()
    st.simulate(paz_remove=paz,paz_simulate=cornFreq2Paz(0.01),remove_sensitivity=True,simulate_sensitivity=True,nfft_pow2=True)
    st.detrend(type="demean") #remover média
    st.detrend(type="linear") #tendêndia linear
    st.taper(type='hamming',max_percentage=0.02) #taper
    st.filter('lowpass',freq=4,zerophase=True) #filtro passa-baixa 4 hz
    st.filter('highpass',freq=0.05,zerophase=True)  #filtro passa-alta 0.05 hz
    st.decimate(10) #reamostragem para 1 hz
    st_orig.plot()
    st.plot()
```

## Looking for events  > 7 mb


```python
event_info = np.genfromtxt('/home/diogo/dados_doutorado/ribeira_process/data_processed/NEIC_2012_2013/NEIC_2012.csv',delimiter=',',skip_header=1,usecols=[1,2,3,4])

#Guardando as variáves cada evento:

ev_time =  np.genfromtxt('/home/diogo/dados_doutorado/ribeira_process/data_processed/NEIC_2012_2013/NEIC_2012.csv',delimiter=',',dtype='str',skip_header=1,usecols=[0])

ev_timeUTC = []
ev_year = []
ev_month = []
ev_day = []
ev_julday = []
ev_hour = []
ev_minute = []
ev_second = []
ev_microsecond = []

for i in ev_time:
	temp = op.UTCDateTime(i)
	ev_year.append(str(temp.year))
	ev_month.append(str(temp.month))
	ev_julday.append(str(temp.julday))
	ev_day.append(str(temp.day))
	ev_hour.append(str(temp.hour))
	ev_minute.append(str(temp.minute))
	ev_second.append(str(temp.second))
	ev_microsecond.append(str(temp.microsecond))
	ev_timeUTC.append(op.UTCDateTime(i))

troca_list = [str(i) for i in np.arange(-1,10,1)]

for i,j in enumerate(ev_day):
	if j in troca_list:
		ev_day[i] = '0'+j 

for i,j in enumerate(ev_month):
	if j in troca_list:
		ev_month[i] = '0'+j 

for i,j in enumerate(ev_julday):
	if j in troca_list:
		ev_julday[i] = '0'+j 

for i,j in enumerate(ev_hour):
	if j in troca_list:
		ev_hour[i] = '0'+j 

for i,j in enumerate(ev_minute):
	if j in troca_list:
		ev_minute[i] = '0'+j 

for i,j in enumerate(ev_second):
	if j in troca_list:
		ev_second[i] = '0'+j 

for i,j in enumerate(ev_microsecond):
	if j == '0':
		ev_microsecond[i] = j+'00'

ev_lat = event_info[:,0]
ev_long = event_info[:,1]
ev_depth = event_info[:,2]
ev_mag = event_info[:,3]
```
