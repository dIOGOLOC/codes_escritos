# Metodologia de Ekstrom et al. 2006

We quantify the agreement between the predicted and observed waveforms using the residual normalized variance (misfit) and the correlation. The misfit F is calculated as:

# <center>$F = \frac{\sum_{i=1}^{n}  (o_{i} - s_{i})^{2}}{\sum_{i=1}^{n}  o_{i}^{2}}$</center>

```python
data_len = 35000
```

```python
F_ROSB = []
for i,j in enumerate(ev_ROSB):
    s_data = syn_ROSB[i].data[:data_len]
    o_data = ev_ROSB[i].data[:data_len]*10**-9
    F_ROSB.append(np.sum(np.power(np.subtract(o_data,s_data),2))/np.sum(np.power(o_data,2)))
    #a = correlate(o_data,s_data,350000)
    #F_ROSB.append(xcorr_max(a))
```

```python
F_ROSB
```

The correlation  C is:

# <center>$C = \frac{\sum_{i=1}^{n}  o_{i} . s_{i}}{\sqrt{\sum_{i=1}^{n}  o_{i}^{2} . \sum_{i=1}^{n}  s_{i}^{2}}}$</center>

```python
C_ROSB = []
for i,j in enumerate(ev_ROSB):
    s_data = syn_ROSB[i].data[:data_len]
    o_data = ev_ROSB[i].data[:data_len]*10**-9
    C_ROSB.append(np.sum(np.multiply(o_data,s_data))/np.sqrt(np.multiply(np.sum(np.power(o_data,2)),np.sum(np.power(s_data,2)))))
```

```python
C_ROSB
```

A third parameter considered is the scaling factor S, which is the factor by which the synthetic seismogram should be multiplied in order to achieve the smallest misfi t:
    
# <center>$S = \frac{\sum_{i=1}^{n}  o_{i} . s_{i}}{\sum_{i=1}^{n}  s_{i}^{2}}$</center>

```python
S_ROSB = []
for i,j in enumerate(ev_ROSB):
    s_data = syn_ROSB[i].data[:data_len]*10**9
    o_data = ev_ROSB[i].data[:data_len]
    S_ROSB.append(np.sum(np.multiply(o_data,s_data))/(np.sum(np.power(s_data,2))))
```

```python
S_ROSB
```

Calculating misfit factor after use the scale factor:

```python
F_ROSB_s = []
for i,j in enumerate(ev_ROSB):
    s_data = syn_ROSB[i].data[:data_len]*S_ROSB[i]
    o_data = ev_ROSB[i].data[:data_len]*10**-9
    F_ROSB_s.append(np.sum(np.power(np.subtract(o_data,s_data),2))/np.sum(np.power(o_data,2)))
```
