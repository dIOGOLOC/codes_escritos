# Filtrando os dados obtidos para um dado intervalo de frequência

```python
freqmin=0.05
freqmax=5
```

```python
for i,j in enumerate(ev):
        j.detrend("demean")
        j.taper(0.05)
        j.filter("bandpass", freqmin=freqmin, freqmax=freqmax)
        j.decimate(factor=10, strict_length=False)
```

```python
for i,j in enumerate(ev_notROT):
        j.detrend("demean")
        j.taper(0.05)
        j.filter("bandpass", freqmin=freqmin, freqmax=freqmax)
        j.decimate(factor=10, strict_length=False)
```

```python
for i,j in enumerate(syn):
        j.detrend("demean")
        j.taper(0.05)
        j.filter("bandpass", freqmin=freqmin, freqmax=freqmax)
        j.decimate(factor=10, strict_length=False)
```

```python
baz_lst = np.arange(180,270.1,0.1)
```

```python
len(baz_lst)
```

```python
ev_notROT_flat_list = [item for sublist in ev_notROT for item in sublist]
```

```python
%%time
ev_notE = ev_notROT_flat_list[::3]
ev_notN = ev_notROT_flat_list[1::3]
ev_notZ = ev_notROT_flat_list[2::3]

R = [[]]*len(ev_notE)
T = [[]]*len(ev_notE)
for k,l in enumerate(ev_notE):
    data = [obspy.signal.rotate.rotate_ne_rt(ev_N[k].data, l.data,  j) for j in baz_lst]
    R[k] = [i[0] for i in data]
    T[k] = [i[1] for i in data]
```

```python
%time 
cc_R = [[]]*len(ev_notE)
for i,j in enumerate(R):
    cc_R[i] = [correlate(syn_N[i], l, 3500) for l in j] 
```

```python
shift_value_R = [[]]*len(ev_notE)
for i,j in enumerate(cc_R):
    shift_value_R[i] = [xcorr_max(k)[1] for k in j]
```

```python
idx  = [[]]*len(ev_notE)
max_R_value  = [[]]*len(ev_notE)

for i,j in enumerate(shift_value_R):
    idx[i] = [k for k,l in enumerate(j) if l == max(j)]
    max_R_value[i] = [l for k,l in enumerate(j) if l == abs(max(j))]
```

```python
f, axarr = plt.subplots(figsize=(20, 30),nrows=len(shift_value_R), ncols=1,sharex=True)
for i,j in enumerate(orglist):
    for r,t in enumerate(shift_value_R[j]):
        axarr[i].plot(baz_lst[r],abs(t),'ok')
    axarr[i].plot(ev_N[j].stats.sac.baz,abs(shift_value_N[j]),'or')
    axarr[i].plot(baz_lst[idx[j]],max_R_value[j],'og')
    for k,l in enumerate(max_R_value[j]):
        axarr[i].text(baz_lst[idx[j]][k],abs(l),str(baz_lst[idx[j]][k]))
    axarr[i].set_ylim(-1,1)
    axarr[i].set_title(station[i])
    #axarr[i].ylabel('Cross-Correlation')
```

```python
shift_value_T = [[]]*len(ev_notE)
for i,j in enumerate(cc_T):
    shift_value_T[i] = [xcorr_max(k)[1] for k in j]
```

```python
idx_T  = [[]]*len(ev_notE)
max_T_value  = [[]]*len(ev_notE)

for i,j in enumerate(shift_value_T):
    idx_T[i] = [k for k,l in enumerate(j) if abs(l) == abs(max(j))]
    max_T_value[i] = [l for k,l in enumerate(j) if abs(l) == abs(max(j))]
```
